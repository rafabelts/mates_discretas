## Operaciones
Python proporciona operadores y métodos para realizar ooperaciones básicas de conjuntos como unión, intersección, diferencia y diferencia simétrica.

### Unión
La unión de dos conjuntos A y B es el conjunto A U B formado por los elementos de A y los Elementos de B es decir:

$$
A \cup B = {x:x \in A \quad o \quad x \in B}
$$


```python
A | B
A.union(B)
```

```python
from matplotlib_venn import venn2
from matplotlib import pyplot as plt

venn2([A, B], (A, B))
plt.show()
```
